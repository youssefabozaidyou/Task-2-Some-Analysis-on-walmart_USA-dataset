```python
import pandas as pd

df = pd.read_excel(r"C:\Users\youss\Desktop\walmart_USA.xlsx")
df = pd.DataFrame(df)
df
```

```python
df.info()
```

```python
print(df.isna().sum().sum())
```

```python
print(df.groupby('Product Name')['Sales'].max())
```

```python
print(df.groupby('Product Name')['Profit'].max())
```

```python
print(df['City'].value_counts())
```

```python
print(df['Category'].value_counts())
```

```python
for i in df.index:
    if df.loc[i, 'Profit'] <= 0 :
        df.drop(i, inplace=True)
df
```

```python
df.describe()
```

```python
print(df['Country'].value_counts())
```

```python
print(df.groupby('Country')['Sales'].sum())
```

```python
df['order_data'] = df['Ship Date'] - df['Order Date']
print(df.groupby('Product Name')['order_data'].count())
```

```python
df
```

```python
df['Unit Price'] = df['Sales'] / df['Quantity']
df
```

```python
for i in df.index:
    if df.loc[i, 'order_data'].days > 5:
        print(df.loc[i])
print('-------------------------------')
df[df['order_data'].dt.days > 5]
```
